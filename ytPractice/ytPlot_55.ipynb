```python
import yt
import numpy as np
```

```python
ds = yt.load("plt_SF_prim_mag000550000")
```

```python
ds.field_list
```

```python
slc = yt.SlicePlot(ds,'z', 'struct_fact_rho_Temp')
slc.set_cmap(field=("struct_fact_rho_Temp"), cmap="turbo")
slc.show()
```

```python
sc = yt.create_scene(ds, field = 'struct_fact_rho_Temp', lens_type="perspective")
source = sc[0]
bounds = (1e-25, 1e-20)
tf = yt.ColorTransferFunction(np.log10(bounds))

tf.add_layers(5, colormap="Waves")

source.tfh.tf = tf
source.tfh.bounds = bounds
sc.show()
```
